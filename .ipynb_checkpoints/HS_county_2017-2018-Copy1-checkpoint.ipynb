```python
import pandas as pd
import csv
import numpy as np
import matplotlib.pyplot as plt
```

```python
students_county= pd.read_csv("county_grade.csv")
students_county.head()

```

```python
HS_county = students_county[['9','10','11','12']]
HS_county = HS_county.apply(pd.to_numeric)
HS_county_join = students_county[['County']].join(HS_county)
HS_county_join
```

```python
#This is the code to generate columns calculating the percentages of the population in AZ based on data on
# https://www.hhs.gov/ash/oah/facts-and-stats/national-and-state-data-sheets/
# adolescents-and-substance-abuse/arizona/index.html
HS_county_join['Totals'] = HS_county[HS_county.columns].sum(axis=1)
HS_county_join
ratio_pop =[]
#smoked = []
#alcohol_before_13 = []
a_drink = []
more_drinks = []
cocaine = []
for i in range(len(HS_county_join['Totals'])):
    ratio_pop.append((HS_county_join.iloc[i, 5 ]/331830)*100)
    #smoked.append((HS_county_join.iloc[i, 5 ]*.30))
    #alcohol_before_13.append((HS_county_join.iloc[i, 5 ]*.18))
    a_drink.append((HS_county_join.iloc[i, 5 ]*.33))
    more_drinks.append((HS_county_join.iloc[i, 5 ]*.18))
    cocaine.append((HS_county_join.iloc[i, 5 ]*.06))
    next
    
HS_county_join['% Students'] = ratio_pop
#HS_county_join['30% Smoked'] = smoked  
#HS_county_join['18% Drank Alcohol Before 13'] = alcohol_before_13
HS_county_join['33% 1 Drink In 30 days'] = a_drink
HS_county_join['18% 4+ Drinks in 30 days'] = more_drinks
HS_county_join['6% Used Cocaine'] = cocaine
```
