```python
%matplotlib inline
import pandas as pd
import csv
import numpy
import matplotlib.pyplot as plt
county_color={}
```

```python
students_county2016= pd.read_csv("county_grade2016.csv")
students_county2016.head()

```

```python
color_scheme = {'Gold':'#FFD700', 'Light Sky Blue':'#87CEFA', 'Light Coral':'#F08080'}
county_color = {'9': color_scheme['Gold'], '10': color_scheme['Light Sky Blue'], '11': color_scheme['Light Coral']}
HS_county2016 = students_county2016[['9','10','11','12']]
HS_county2016 = HS_county2016.apply(pd.to_numeric)
HS_county2016_join = students_county2016[['County']].join(HS_county2016)
HS_county2016_join
```

```python
HS_county2016_join['Totals'] = HS_county2016[HS_county2016.columns].sum(axis=1)
HS_county2016_join
```

```python
plt.scatter(HS_county2016_join['County'], HS_county2016_join['9'],s=HS_county2016_join['9'],color = county_color['9'], edgecolor = 'black', label = 'County',alpha=0.75)
#plt.scatter(HS_county2016_join['County'], HS_county2016_join['10'])
plt.title('County Vs Drug Use')
plt.xlabel('County')
plt.ylabel('Drug Use')
plt.show

#plt.scatter(HS_county2016_join['County'], HS_county2016_join['Totals'], s=HS_county2016_join['Totals'], color = HS_county2016_join['Totals'], edgecolor = 'black', label = 'Rural', alpha = .75)
```
