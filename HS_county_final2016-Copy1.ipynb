```python
%matplotlib inline
import pandas as pd
import csv
import numpy as np
import matplotlib.pyplot as plt
county_color={}
```

```python
students_county2016= pd.read_csv("county_grade2016totals.csv")
students_county2016.head()

```

```python
#color_scheme = {'Gold':'#FFD700', 'Light Sky Blue':'#87CEFA', 'Light Coral':'#F08080'}
#county_color = {'9': color_scheme['Gold'], '10': color_scheme['Light Sky Blue'], '11': color_scheme['Light Coral']}
```

```python
HS_county2016 = students_county2016[['8','10','12']]
HS_county2016 = HS_county2016.apply(pd.to_numeric)
HS_county_join2016 = students_county2016[['County']].join(HS_county2016)
HS_county_join2016
```

```python
HS_county_join2016['Totals'] = HS_county2016[HS_county2016.columns].sum(axis=1)
HS_county_join2016
```

```python
#Lists
mj_8th = []
mj_10th = []
mj_12th = []
alcohol_8th = []
alcohol_10th = []
alcohol_12th = []
e_cig_8th = []
e_cig_10th = []
e_cig_12th = []
```

```python
for i in range(len(HS_county_join2016['Totals'])):
 
    mj_8th.append((HS_county_join2016.iloc[i, 1 ]*.054))
    mj_10th.append((HS_county_join2016.iloc[i, 2 ]*.14))
    mj_12th.append((HS_county_join2016.iloc[i, 3 ]*.225))
    alcohol_8th.append((HS_county_join2016.iloc[i, 1 ]*.073))
    alcohol_10th.append((HS_county_join2016.iloc[i, 2 ]*.332))
    alcohol_12th.append((HS_county_join2016.iloc[i, 3 ]*.332))
    e_cig_8th.append((HS_county_join2016.iloc[i, 1 ]*.062))
    e_cig_10th.append((HS_county_join2016.iloc[i, 2 ]*.103))
    e_cig_12th.append((HS_county_join2016.iloc[i, 3 ]*.124))

    
HS_county_join2016['Marijuana 8th grade'] = mj_8th
HS_county_join2016['Marijuana 10th grade'] = mj_10th    
HS_county_join2016['Marijuana 12th grade'] = mj_12th
HS_county_join2016['Alcohol 8th grade'] = alcohol_8th
HS_county_join2016['Alcohol 10th grade'] = alcohol_10th
HS_county_join2016['Alcohol 12th grade'] = alcohol_12th
HS_county_join2016['E-Cigarettes 8th grade'] = e_cig_8th
HS_county_join2016['E-Cigarettes 10th grade'] = e_cig_10th
HS_county_join2016['E-Cigarettes 12th grade'] = e_cig_12th
HS_county_join2016
```

```python
fig = plt.figure(figsize=(10,8))
barWidth = 0.3

mj = [4660,11933,12219]
alcohol = [6300,28299,28360]
e_cig=[5350,8779,10592]
bars1 = ('8th Grade', '10th Grade','12th Grade')

y1 = np.arange(len(bars1))
y2= [x + barWidth for x in y1]
y3= [x + barWidth for x in y2]
 
# Create bars
plt.bar(y1, mj, width = barWidth, color = (0.3,0.1,0.4,0.6), edgecolor='black', label='Marijuana')
plt.bar(y2, alcohol, width = barWidth, color = (0.3,0.5,0.4,0.6), edgecolor='black', label='Alcohol')
plt.bar(y3, e_cig, width = barWidth, color = (0.1,0.1,0.3,0.5), edgecolor='black', label='Ecigarettes')
 
# Create names on the x-axis
plt.xticks([y + (.95*barWidth) for y in range(len(bars1))], bars1)
plt.title('Substance abuse in AZ: Age 12-17 (National Average)')
plt.xlabel('Substance')
plt.ylabel('Population')
plt.legend()

#Save Figure
fig.savefig('Substance abuse in AZ: Age 12-17 (National Average) .png') 
# Show graphic
plt.show()
```
