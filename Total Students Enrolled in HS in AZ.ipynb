```python
import pandas as pd
import csv
import numpy
```

```python
population_students= pd.read_csv("population_grade.csv")
population_students.head()
```

```python
ninth_twelft_grade = population_students[['9','10','11','12']]
total_grades = ninth_twelft_grade.replace('*','0')

total_grades = total_grades[(total_grades[['9','10','11','12']] != '0').all(axis=1)]
total_grades = total_grades.apply(pd.to_numeric)
```

```python
total_students_grade = total_grades.sum()
```

```python
pd.DataFrame(total_students_grade, columns=['Total'])
    

```
